```python
import pandas as pd
path="melb_data.csv"
data=pd.read_csv(path)
data.head()
```

```python
y=data.Price
```

```python
melb_features=data.drop(['Price'], axis=1)
X=melb_features.select_dtypes(exclude=[object])
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid=train_test_split(X,y, train_size=0.8, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
def selection(X_train, X_valid, y_train, y_valid):
    model=RandomForestRegressor()
    model.fit(X_train,y_train)
    predections=model.predict(X_valid)
    return mean_absolute_error(predections,y_valid)
```

### column dropping techniques

```python
col_missing_data=[
    col for col in X_train.columns
    if X_train[col].isnull().any()
]
col_missing_data
```

```python
Dropped_X_train=X_train.drop(col_missing_data, axis=1)
Dropped_X_valid=X_valid.drop(col_missing_data, axis=1)
```

```python
Dropped_X_train
```

```python
print("MAE from dropping table is :", selection(Dropped_X_train,Dropped_X_valid, y_train, y_valid))
```

### from imputation

```python
from sklearn.impute import SimpleImputer
myImputer=SimpleImputer()
imputed_X_train=pd.DataFrame(myImputer.fit_transform(X_train))
```

```python
imputed_X_valid=pd.DataFrame(myImputer.fit_transform(X_valid))
```

```python
print("MAE from dropping table is :", selection(Dropped_X_train,Dropped_X_valid, y_train, y_valid))
print("MAE from imputing table is :", selection(imputed_X_train,imputed_X_valid, y_train, y_valid))
```

### extension of imputing:

```python
X_train_new=X_train.copy()
X_valid_new=X_valid.copy()
```

```python
for col in col_missing_data:
    X_train_new[col+'_was missing']=X_train_new[col].isnull()
    X_valid_new[col+'_was missing']=X_valid_new[col].isnull()
```

```python
imputed_X_train_new=pd.DataFrame(myImputer.fit_transform(X_train_new))
imputed_X_valid_new=pd.DataFrame(myImputer.fit_transform(X_valid_new))
imputed_X_train_new.columns=X_train_new.columns
imputed_X_valid_new.columns=X_valid_new.columns
```

```python
print("MAE from dropping table is :", selection(Dropped_X_train,Dropped_X_valid, y_train, y_valid))
print("MAE from imputing table is :", selection(imputed_X_train,imputed_X_valid, y_train, y_valid))
print("MAE from extension imputing table is :", selection(imputed_X_train_new,imputed_X_valid_new, y_train, y_valid))
```
